```python
import pandas as pd
import matplotlib.pyplot as plt
# import os
# os.getcwd()
```

```python
df = pd.read_csv("../traj.csv")

plt.figure()
plt.plot(df["px"], df["pz"])
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.title("XY Trajectory")
plt.axis("equal")
plt.grid()
plt.show()

plt.figure()
plt.plot(df["t"], df["pz"])
plt.xlabel("time [s]")
plt.ylabel("z [m]")
plt.title("Altitude")
plt.grid()
plt.show()
```
