```python
import numpy as np
import matplotlib.pyplot as plt

# 시스템: dx/dt = -x + u, 초기조건 x(0)=5
def system(x, u):
    return -x + u

dt = 0.01
T = 5
time = np.arange(0, T, dt)

x = np.zeros_like(time)
u = np.ones_like(time) * 2  # 제어입력: 상수 2
x[0] = 5

for i in range(1, len(time)):
    x[i] = x[i-1] + dt * system(x[i-1], u[i-1])

plt.plot(time, x, label="x(t)")
plt.plot(time, u, '--', label="u(t)")
plt.xlabel("Time [s]")
plt.legend()
plt.title("Simple State Equation Simulation")
plt.show()
```
