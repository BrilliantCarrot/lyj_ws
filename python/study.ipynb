```python
class Person:
    def __init__(self, name, age, address,wallet):
        self.hello = "안녕하세요"
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet # 변수 앞에 __를 붙여서 비공개(private) 속성으로 만듦
        # __이름__의 경우엔 공개(public)
    def greeting(self):
        print("{0} 저는 {1}입니다.".format(self.hello, self.name))
    def pay(self, amount):
        self.__wallet -= amount
        print("이제 {0}원 남았네요".format(self.__wallet))

maria = Person("마리아",20,"서울시 서초구 반포동")
maria.greeting()
print('이름:', maria.name)
print('나이:', maria.age)
print('주소:', maria.address)
```

```python
class Person:
    bag = []
    def put_bag(self, stuff):
        self.bag.append(stuff)
 
james = Person()
james.put_bag('책')
 
maria = Person()
maria.put_bag('열쇠')
 
print(james.bag)
print(maria.bag)
```

```python
class Calc:
    @staticmethod
    def add(a, b):
        print(a + b)
 
    @staticmethod
    def mul(a, b):
        print(a * b)
# 정적 메서드는 self를 받지 않으므로 인스턴스 속성에 접근할 수 없다.
# 그래서 보통 정적 메서드는 인스턴스 속성, 인스턴스 메서드가 필요 없을 때 사용한다.
# 정적 메서드는 메서드의 실행이 외부 상태에 영향을 끼치지 않는 순수 함수를 만들 때 사용.
Calc.add(10, 20)    # 클래스에서 바로 메서드 호출
Calc.mul(10, 20)    # 클래스에서 바로 메서드 호출
```

```python
class Person:
    count = 0    # 클래스 속성
    def __init__(self):
        Person.count += 1    # 인스턴스가 만들어질 때
                             # 클래스 속성 count에 1을 더함
    @classmethod
    def print_count(cls):
        print('{0}명 생성되었습니다.'.format(cls.count))    # cls로 클래스 속성에 접근
 
james = Person()
maria = Person()
Person.print_count()    # 2명 생성되었습니다.
```
