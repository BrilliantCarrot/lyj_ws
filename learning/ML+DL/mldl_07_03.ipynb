```python
from tensorflow import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = \
    keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
# import pandas as pd
# california = pd.read_csv("./sample_data/california_housing_test.csv",encoding="utf-8")
# california.head()
```

```python
def model_fn(a_layer=None):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(28, 28)))
    model.add(keras.layers.Dense(100, activation='relu'))
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dense(10, activation='softmax'))    # 다중 분류
    return model
model = model_fn()
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=5, verbose=2)
```

모델의 검증 점수가 상승하는것을 2번씩 작업하지 않게 하는 callbacks 매개변수  
ModelCheckpoint에 최상의 검증 점수를 만드는 모델이 저장됨

```python
def model_fn(a_layer=None):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(28, 28)))
    model.add(keras.layers.Dense(100, activation='relu'))
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dropout(0.3))    # 드롭 아웃도 기본적으로 추가한 버전 model_fn
    model.add(keras.layers.Dense(10, activation='softmax'))    # 다중 분류
    return model
model = model_fn()
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')

# 최상의 검증 점수를 만드는 모델을 "best-model.h5로 저장"
checkpoint_cb = keras.callbacks.ModelCheckpoint("best-model.h5", save_best_only=True)

model.fit(train_scaled, train_target, epochs=20, verbose=2, validation_data=(val_scaled, val_target),
          callbacks=[checkpoint_cb])
```

```python
model = keras.models.load_model('./drive/MyDrive/개인 공부/ML_DL/data/best-model.h5')
model.evaluate(val_scaled, val_target)
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')

checkpoint_cb = keras.callbacks.ModelCheckpoint('./drive/MyDrive/개인 공부/ML_DL/data/best-model.h5', save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

# 20번의 에포크를 지정하지만 조기 종료가 가능
history = model.fit(train_scaled, train_target, epochs=20, verbose=1, validation_data=(val_scaled, val_target),
                    callbacks=[checkpoint_cb, early_stopping_cb])
```
